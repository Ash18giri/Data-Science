
1.User-based collaborative filtering (UBCF) recommends items to users based on similar users, while item-based collaborative filtering (IBCF) recommends items based on similar items

2.Collaborative filtering is a method for recommending items to users based on how other users with similar interests have interacted with those items. It's a popular technique used in recommender systems for music and movies.
How it works
1)Create a user-item matrix: Build a database of user preferences for items.
2)Group users by behavior: Find users with similar preferences and group them together.
3)Calculate similarities: Compare the profiles of users to find similarities between them.
4)Make recommendations: Recommend items to users based on the preferences of similar users.
This can be done through Cosine based similarity or corelation based similarity.
