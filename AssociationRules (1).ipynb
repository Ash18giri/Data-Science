```python
import mlxtend
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel("Online retail.xlsx", header = None)

dg = df[0].str.split(',', expand=True)
dg
```

```python
seen_values = set()
all_unique_values = []  # Store all unique values across columns

for column in dg.columns:
    unique_values = dg[column].unique()

    # Filter out previously seen values
    filtered_unique_values = [val for val in unique_values if val not in seen_values and val is not None]

    # Update seen values with new unique values
    seen_values.update(filtered_unique_values)

    print(f"Unique values in column {column}: {filtered_unique_values}")

    all_unique_values.extend(filtered_unique_values)  # Add unique values to the list

# Print all unique values found across all columns
print("All unique values:", all_unique_values)
```

```python
# Create a new DataFrame with columns as all unique values, initialized to 0
one_hot_df = pd.DataFrame(0, index=dg.index, columns=all_unique_values)

# Iterate over each row and update the one-hot encoded values
for index, row in dg.iterrows():
    present_items = row.dropna().unique()  # Get UNIQUE non-null items in the row
    one_hot_df.loc[index, present_items] = 1  # Set 1 where item is present

# Display the first few rows of the transformed dataset
one_hot_df
```

```python
# Apply Apriori Algorithm to find frequent item sets
frequent_itemsets = apriori(one_hot_df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
# Generate association rules based on confidence
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7)
rules
```

```python
# Filter rules based on lift and confidence thresholds
filtered_rules = rules[(rules['lift'] > 1) & (rules['confidence'] > 0.2)]

# Display the association rules
print(filtered_rules)
```

1. Key Metrics for Rule Analysis
Each association rule contains:

Antecedents: Items that were purchased (X in X â†’ Y).
Consequents: Items that are likely to be purchased (Y in X â†’ Y).
Support: How frequently the itemset appears in transactions.
Confidence: The probability that Y is purchased when X is purchased.
Lift: How much more likely Y is purchased when X is purchased, compared to random chance.

1.   In association rule mining, "lift" is a metric that measures the strength of the relationship between two items in a rule, indicating how much more likely the consequent item is to occur when the antecedent item is present, compared to if they were statistically independent; essentially, it tells you how much stronger the association is than what you would expect by chance, making it crucial for identifying truly meaningful and actionable relationships within your data.

Importance in Association Rules:
Filtering out spurious rules: While a rule may have high confidence, a low lift value might suggest the relationship is not as strong as it appears, potentially due to the high popularity of the items involved.
Prioritizing actionable insights: By focusing on rules with high lift values, businesses can identify the most valuable relationships between products or events to inform marketing strategies and product placement decisions.



2. In association rule mining, "support" measures how frequently a specific itemset appears in the dataset, while "confidence" indicates the likelihood of an item appearing in a transaction given that another specific item is already present in that transaction; essentially, support tells you how common an itemset is, and confidence tells you how strong the association between two items is within a rule.

Calculation:

Support:
Formula: Support(X) = (Number of transactions containing itemset X) / (Total number of transactions)
Explanation: This calculates the proportion of transactions where a specific set of items (X) appears.

Confidence:
Formula: Confidence(X -> Y) = Support(X âˆª Y) / Support(X)
Explanation: This calculates the probability of itemset Y appearing in a transaction given that itemset X is already present.

3. Some key limitations and challenges of association rules mining include: generating a large number of potentially irrelevant rules, difficulty in choosing appropriate thresholds (support and confidence), handling noisy data, interpreting the meaning of discovered rules, scalability issues with large datasets, and the potential for generating "obvious" or uninteresting rules; making it crucial to carefully select the right metrics and filter results to identify truly valuable associations.

Key points about these limitations:

1. Overwhelming rule generation:
Association rule mining can produce a huge number of rules, making it difficult to manually analyze and interpret which ones are truly significant.
2. Threshold selection:
Choosing appropriate minimum support and confidence levels is critical, as setting them too low can generate too many irrelevant rules, while setting them too high might miss important patterns.
3. Data quality issues:
Noise and outliers in the data can significantly affect the quality of mined association rules.
4. Interpretability:
While rules may be statistically significant, they might not always have clear practical interpretations in real-world contexts.
5. Scalability concerns:
For large datasets, the computational cost of association rule mining can be high, requiring optimization techniques.
6. "Obvious" rules:
The algorithm may sometimes identify rules that are already well-known or considered common sense, lacking actionable insight
