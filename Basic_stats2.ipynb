```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
s=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
df=pd.DataFrame(s,columns=['print_head'])
df
```

```python
df.describe()
```

```python
from scipy import stats
print_head_ci=stats.t.interval(0.99,len(df)-1,df['print_head'].mean(),df['print_head'].std())
print(np.round(print_head_ci,4))
```
